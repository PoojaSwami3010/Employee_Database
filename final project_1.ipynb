```python
#  2.Random Password Generator :Password must have capital,small,number,symbol,
#leng 8 plus generate 25 password forabove employee?
# random password generator code
import string
import random

emp_password=[]
def random_password(string_length):
        password_char=string.ascii_letters+string.digits+string.punctuation
        emp_password.append(''.join(random.choice(password_char) for i in range(string_length)))
        return emp_password
    
i=1
while i<26:
    string_length=int(input("enter string length: "))
    random_password(string_length)
    i+=1
    if i==26:
        break
```
